```python
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity

model = SentenceTransformer('sentence-transformers/all-MiniLM-L6-v2')

cv_text = """
Experienced software engineer with 3 years of experience in backend development using Python, 
Django, and REST APIs. Skilled in database design, cloud deployment (AWS), and team collaboration. 
Looking for opportunities to contribute to scalable web applications.
"""

job_description = """
We are hiring a backend developer proficient in Python, Django, and RESTful APIs.
Experience with cloud infrastructure (AWS or Azure) and database optimization is a plus.
"""

cv_embedding = model.encode([cv_text])
job_embedding = model.encode([job_description])

similarity_score = cosine_similarity(cv_embedding, job_embedding)[0][0]

print(f"Semantic Similarity Score: {similarity_score:.4f}")

if similarity_score > 0.8:
    print("Very Strong Match — Candidate highly fits the job.")
elif similarity_score > 0.6:
    print("Moderate Match — Candidate is somewhat suitable.")
else:
    print("Weak Match — Candidate likely not suitable.")
```
